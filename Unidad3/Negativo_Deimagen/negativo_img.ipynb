```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def constraste(img, factor=1.0):
    img=img.astype(np.float32)
    for i in range(3):
        media= np.mean(img[:,:,i])
        img[:,:,i]= media+(img[:,:,i] - media) * factor
        img[:,:,i]= np.clip(img[:,:,i], 0, 255)
    
    return img.astype(np.uint8)
img= cv2.imread('goku.jpg')
contrast=3
img_fin= constraste(img, contrast)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img_fin, cv2.COLOR_BGR2RGB))
plt.title('contraste')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def brillo (img, brillo=0):
    img_brillo= img.astype(np.float32) + brillo
    img_brillo= np.clip(img_brillo, 0, 255).astype(np.uint8)
    return img_brillo
img= cv2.imread('goku.jpg', 0)
bril=-70
img_fin= brillo(img, bril)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(img_fin,cmap='gray')
plt.title('brillo')
plt.show()
```
