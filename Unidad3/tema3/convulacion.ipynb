Convoluciones con matriz kernel

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread('goku.jpg')
kernel=np.array([(0,-1,0),(-1,5,-1),(0,-1,0)])#mascara de convolucion
img_fin=cv2.filter2D(img,-1,kernel)

plt.subplot(121)
plt.imshow(img,cmap='gray')
plt.subplot(122)
plt.imshow(img_fin,cmap='gray')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('goku.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  

kernel = np.ones((5,5), np.float32) / 25
suavizada = cv2.filter2D(img, -1, kernel)
kernel2 = np.ones((7,7), np.float32) / 49
suavizada2 = cv2.filter2D(img, -1, kernel2)

plt.subplot(131)
plt.imshow(img)  
plt.title('Original')
plt.subplot(132)
plt.imshow(suavizada)
plt.title('Kernel 5x5')
plt.subplot(133)
plt.imshow(suavizada2)
plt.title('Kernel 7x7')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('goku.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Definir kernels
k1 = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
k2 = np.ones((5, 5), np.float32) / 25
k3 = np.array([[0, 0, 0], [-1, 1, 0], [0, 0, 0]])
k4 = np.array([[-2, -1, 0], [-1, 1, 1], [0, 1, 2]])
k5 = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])
k6 = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
k7 = np.array([[1, -2, 1], [-2, 5, -2], [1, -2, 1]])
k8 = np.array([[1, 1, 1], [1, -2, 1], [-1, -1, -1]])
k9 = np.array([[-1, 1, 1], [-1, -2, 1], [-1, 1, 1]])
k10 = np.array([[1, 2, 3, 1, 1], [2, 7, 11, 7, 2], [3, 11, 17, 11, 3], [2, 7, 11, 7,1], [1, 2, 3, 2, 1]], np.float32) / 256

# ESCALA DE GRISES
plt.figure(figsize=(24, 8))
plt.suptitle('11 FILTROS - ESCALA DE GRISES', fontsize=16, fontweight='bold')

plt.subplot(2, 6, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(2, 6, 2)
plt.imshow(cv2.filter2D(img, -1, k1), cmap='gray')
plt.title('Enfoque')
plt.axis('off')

plt.subplot(2, 6, 3)
plt.imshow(cv2.filter2D(img, -1, k2), cmap='gray')
plt.title('Desenfoque')
plt.axis('off')

plt.subplot(2, 6, 4)
plt.imshow(cv2.filter2D(img, -1, k3), cmap='gray')
plt.title('Realce')
plt.axis('off')

plt.subplot(2, 6, 5)
plt.imshow(cv2.filter2D(img, -1, k4), cmap='gray')
plt.title('Repujado')
plt.axis('off')

plt.subplot(2, 6, 6)
plt.imshow(cv2.filter2D(img, -1, k10), cmap='gray')
plt.title('Gauss')
plt.axis('off')

plt.subplot(2, 6, 7)
plt.imshow(cv2.filter2D(img, -1, k5), cmap='gray')
plt.title('Bordes')
plt.axis('off')

plt.subplot(2, 6, 8)
plt.imshow(cv2.filter2D(img, -1, k6), cmap='gray')
plt.title('Sobel')
plt.axis('off')

plt.subplot(2, 6, 9)
plt.imshow(cv2.filter2D(img, -1, k7), cmap='gray')
plt.title('Sharpen')
plt.axis('off')

plt.subplot(2, 6, 10)
plt.imshow(cv2.filter2D(img, -1, k8), cmap='gray')
plt.title('Norte')
plt.axis('off')

plt.subplot(2, 6, 11)
plt.imshow(cv2.filter2D(img, -1, k9), cmap='gray')
plt.title('Este')
plt.axis('off')

plt.tight_layout()
plt.show()

# A COLORES
plt.figure(figsize=(24, 8))
plt.suptitle('11 FILTROS - A COLORES (RGB)', fontsize=16, fontweight='bold')

plt.subplot(2, 6, 1)
plt.imshow(img_rgb)
plt.title('Original')
plt.axis('off')

plt.subplot(2, 6, 2)
plt.imshow(np.clip(cv2.filter2D(img_rgb, -1, k1), 0, 255).astype(np.uint8))
plt.title('Enfoque')
plt.axis('off')

plt.subplot(2, 6, 3)
plt.imshow(np.clip(cv2.filter2D(img_rgb, -1, k2), 0, 255).astype(np.uint8))
plt.title('Desenfoque')
plt.axis('off')

plt.subplot(2, 6, 4)
plt.imshow(np.clip(cv2.filter2D(img_rgb, -1, k3), 0, 255).astype(np.uint8))
plt.title('Realce')
plt.axis('off')

plt.subplot(2, 6, 5)
plt.imshow(np.clip(cv2.filter2D(img_rgb, -1, k4), 0, 255).astype(np.uint8))
plt.title('Repujado')
plt.axis('off')

plt.subplot(2, 6, 6)
plt.imshow(np.clip(cv2.filter2D(img_rgb, -1, k10), 0, 255).astype(np.uint8))
plt.title('Gauss')
plt.axis('off')

plt.subplot(2, 6, 7)
plt.imshow(np.clip(cv2.filter2D(img_rgb, -1, k5), 0, 255).astype(np.uint8))
plt.title('Bordes')
plt.axis('off')

plt.subplot(2, 6, 8)
plt.imshow(np.clip(cv2.filter2D(img_rgb, -1, k6), 0, 255).astype(np.uint8))
plt.title('Sobel')
plt.axis('off')

plt.subplot(2, 6, 9)
plt.imshow(np.clip(cv2.filter2D(img_rgb, -1, k7), 0, 255).astype(np.uint8))
plt.title('Sharpen')
plt.axis('off')

plt.subplot(2, 6, 10)
plt.imshow(np.clip(cv2.filter2D(img_rgb, -1, k8), 0, 255).astype(np.uint8))
plt.title('Norte')
plt.axis('off')

plt.subplot(2, 6, 11)
plt.imshow(np.clip(cv2.filter2D(img_rgb, -1, k9), 0, 255).astype(np.uint8))
plt.title('Este')
plt.axis('off')

plt.tight_layout()
plt.show()
```
