# Crear Imagen con Cuadrantes RGB
Crear una imagen de 100x100 píxeles dividida en 4 cuadrantes con colores: Rojo, Verde, Azul y Negro

```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
```

```python
tamanio = 100
mitad = tamanio // 2

imagen = np.zeros((tamanio, tamanio, 3), dtype=np.uint8)

rojo = [0, 0, 255]
verde = [0, 255, 0]
azul = [255, 0, 0]
negro = [0, 0, 0]

cuadrantes = [
    [(0, mitad), (0, mitad), rojo],
    [(0, mitad), (mitad, tamanio), verde],
    [(mitad, tamanio), (0, mitad), azul],
    [(mitad, tamanio), (mitad, tamanio), negro]
]

for (y_inicio, y_fin), (x_inicio, x_fin), color in cuadrantes:
    cuadrante = np.ones((y_fin - y_inicio, x_fin - x_inicio, 3), dtype=np.uint8) * np.array(color)
    imagen[y_inicio:y_fin, x_inicio:x_fin] = cuadrante

print("Imagen creada con numpy.ones + ciclo for")
print(f"Dimensiones: {imagen.shape}")
```

```python
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 10))
plt.imshow(imagen_rgb)
plt.title('Imagen con 4 Cuadrantes RGB', fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
ruta_salida = 'imagen.jpeg'
cv2.imwrite(ruta_salida, imagen)
print(f"Imagen guardada como: {ruta_salida}")
```

```python
colores_rubik = {
    'blanco': [255, 255, 255],
    'amarillo': [0, 255, 255],
    'rojo': [0, 0, 255],
    'naranja': [0, 165, 255],
    'azul': [255, 0, 0],
    'verde': [0, 128, 0]
}

colores_lista = list(colores_rubik.values())

tamanio_celda = 50
filas = 3
columnas = 3
tamanio_rubik = tamanio_celda * filas

rubik = np.zeros((tamanio_rubik, tamanio_rubik, 3), dtype=np.uint8)

np.random.seed()
for i in range(filas):
    for j in range(columnas):
        color_random = colores_lista[np.random.randint(0, len(colores_lista))]
        y_inicio = i * tamanio_celda
        y_fin = y_inicio + tamanio_celda
        x_inicio = j * tamanio_celda
        x_fin = x_inicio + tamanio_celda
        
        celda = np.ones((tamanio_celda, tamanio_celda, 3), dtype=np.uint8) * np.array(color_random)
        rubik[y_inicio:y_fin, x_inicio:x_fin] = celda

print(f"Rubik 3x3 creado con dimensiones: {rubik.shape}")
```

```python
rubik_rgb = cv2.cvtColor(rubik, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8, 8))
plt.imshow(rubik_rgb)
plt.title('Cubo Rubik 3x3 - Colores Aleatorios', fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
cv2.imwrite('rubik_3x3.jpeg', rubik)
print("Rubik guardado como: rubik_3x3.jpeg")
```
