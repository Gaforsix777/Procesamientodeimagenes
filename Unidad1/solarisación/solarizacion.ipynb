```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def posterizar (imagen,niveles):
    factor_escala=255/niveles
    img_pos=np.round(imagen/factor_escala)*factor_escala
    return img_pos.astype(np.uint8)

def solarizar(imagen, umbral):
    #APLICAMOS SOL
    img_sol=np.where(imagen<umbral,imagen,255-imagen)
    return img_sol

imagen=cv2.imread('dragon.jpg')
umbral=127
img_fin=solarizar(imagen, umbral)

nvl=1.5
img_fin2=posterizar(imagen, nvl)

plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')

plt.subplot(132)
plt.imshow(cv2.cvtColor(img_fin, cv2.COLOR_BGR2RGB))
plt.title('Imagen Solarizada')

plt.subplot(133)
plt.imshow(cv2.cvtColor(img_fin2, cv2.COLOR_BGR2RGB))
plt.title('Imagen Posterizada')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def falso_color(img_gray):
    escala_img = cv2.normalize(img_gray, None, 0, 255, cv2.NORM_MINMAX)
    falso_img = cv2.applyColorMap(escala_img, cv2.COLORMAP_JET)
    return falso_img

imagen = cv2.imread('dragon.jpg')
img_gray = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
img_final = falso_color(img_gray)

# Mostrar las imágenes
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
axs[0].set_title('Imagen Original')
axs[0].axis('off')
axs[1].imshow(cv2.cvtColor(img_final, cv2.COLOR_BGR2RGB))
axs[1].set_title('Imagen Falso Color')
axs[1].axis('off')
plt.show()
```
