En este ejercicio aplicamos diferentes esquemas de colorización a una imagen en escala de grises para explorar cómo distintas paletas pueden revelar diferentes características visuales de la imagen original.

con escala de grises

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('dragon.jpg')
img_gray = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

paletas = [
    (cv2.COLORMAP_JET, 'JET'),
    (cv2.COLORMAP_AUTUMN, 'AUTUMN'),
    (cv2.COLORMAP_BONE, 'BONE'),
    (cv2.COLORMAP_HOT, 'HOT'),
    (cv2.COLORMAP_COOL, 'COOL'),
    (cv2.COLORMAP_SPRING, 'SPRING'),
    (cv2.COLORMAP_SUMMER, 'SUMMER'),
    (cv2.COLORMAP_WINTER, 'WINTER'),
    (cv2.COLORMAP_RAINBOW, 'RAINBOW'),
    (cv2.COLORMAP_OCEAN, 'OCEAN'),
    (cv2.COLORMAP_MAGMA, 'MAGMA'),
    (cv2.COLORMAP_INFERNO, 'INFERNO')
]

plt.figure(figsize=(16, 12))

for i, (paleta, nombre) in enumerate(paletas):
    img_coloreada = cv2.applyColorMap(img_gray, paleta)
    plt.subplot(3, 4, i+1)
    plt.imshow(cv2.cvtColor(img_coloreada, cv2.COLOR_BGR2RGB))
    plt.title(nombre)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

imagen = cv2.imread('dragon.jpg')
img_gray = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

print(f"Imagen original shape: {imagen.shape}")
print(f"Imagen escala de grises shape: {img_gray.shape}")

resultado_gray = cv2.applyColorMap(img_gray, cv2.COLORMAP_JET)
resultado_color = cv2.applyColorMap(imagen, cv2.COLORMAP_JET)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(resultado_gray, cv2.COLOR_BGR2RGB))
plt.title('CON escala de grises')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(resultado_color, cv2.COLOR_BGR2RGB))
plt.title('SIN escala de grises')
plt.axis('off')

plt.tight_layout()
plt.show()
```
