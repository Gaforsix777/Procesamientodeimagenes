```python
import cv2
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [4, 3]
img=cv2.imread('imagen.jpg',cv2.IMREAD_COLOR)
#IMAGEN ORIGINAL Y DIVIDIDA POR CANALES

plt.subplot(141)
plt.title("IMAGEN ORIGINAL")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(142)
plt.title("CANAL ROJO")
plt.imshow(img[:,:,0], cmap='gray')

plt.subplot(143)
plt.title("CANAL VERDE")
plt.imshow(img[:,:,1], cmap='gray')

plt.subplot(144)
plt.title("CANAL AZUL")
plt.imshow(img[:,:,2], cmap='gray')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [10, 5]
img=cv2.imread('imagen.jpg',cv2.IMREAD_COLOR)
imgRGB=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

r=imgRGB[:,:,2]
g=imgRGB[:,:,1]
b=imgRGB[:,:,0]

#Crear imagen por canal
r_img=np.zeros_like(imgRGB)
r_img[:,:,0]=r

g_img=np.zeros_like(imgRGB)
g_img[:,:,1]=g

b_img=np.zeros_like(imgRGB)
b_img[:,:,2]=b

plt.figure(figsize=(10,5))
plt.subplot(141)
plt.title("IMAGEN ORIGINAL")
plt.imshow(imgRGB)
plt.axis('off')

plt.subplot(142)
plt.title("CANAL ROJO")
plt.imshow(r_img)
plt.axis('off')

plt.subplot(143)
plt.title("CANAL VERDE")
plt.imshow(g_img)
plt.axis('off')

plt.subplot(144)
plt.title("CANAL AZUL")
plt.imshow(b_img)
plt.axis('off') 

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Cargar imagen
img = cv2.imread('imagen.jpg', cv2.IMREAD_COLOR)
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Crear figura con 11 imagenes (4 columnas x 3 filas)
plt.figure(figsize=(16, 12))

# 1. IMAGEN ORIGINAL
plt.subplot(3, 4, 1)
plt.imshow(imgRGB)
plt.title('IMAGEN ORIGINAL')
plt.axis('off')


plt.subplot(3, 4, 2)
plt.imshow(img_gray, cmap='gray')
plt.title('Colormap: GRAY')
plt.axis('off')


plt.subplot(3, 4, 3)
plt.imshow(img_gray, cmap='viridis')
plt.title('Colormap: VIRIDIS')
plt.axis('off')


plt.subplot(3, 4, 4)
plt.imshow(img_gray, cmap='plasma')
plt.title('Colormap: PLASMA')
plt.axis('off')


plt.subplot(3, 4, 5)
plt.imshow(img_gray, cmap='inferno')
plt.title('Colormap: INFERNO')
plt.axis('off')


plt.subplot(3, 4, 6)
plt.imshow(img_gray, cmap='magma')
plt.title('Colormap: MAGMA')
plt.axis('off')


plt.subplot(3, 4, 7)
plt.imshow(img_gray, cmap='cividis')
plt.title('Colormap: CIVIDIS')
plt.axis('off')


plt.subplot(3, 4, 8)
plt.imshow(img_gray, cmap='bone')
plt.title('Colormap: BONE')
plt.axis('off')


plt.subplot(3, 4, 9)
plt.imshow(img_gray, cmap='cool')
plt.title('Colormap: COOL')
plt.axis('off')


plt.subplot(3, 4, 10)
plt.imshow(img_gray, cmap='hot')
plt.title('Colormap: HOT')
plt.axis('off')


plt.subplot(3, 4, 11)
plt.imshow(img_gray, cmap='spring')
plt.title('Colormap: SPRING')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img= cv2.imread('imagen.jpg')
img_rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

fig, axs = plt.subplots(1, 4, figsize=(20, 5))

axs[0].imshow(img_rgb)
axs[0].set_title('Imagen Original')

fig_img=axs[1].imshow(img_rgb[...,0], cmap='gray')
fig.colorbar(fig_img, ax=axs[1],shrink=0.5)
axs[1].set_title('Canal Rojo')

fig_img=axs[2].imshow(img_rgb[...,1], cmap='gray')
fig.colorbar(fig_img, ax=axs[2],shrink=0.5)
axs[2].set_title('Canal Verde')

fig_img=axs[3].imshow(img_rgb[...,2], cmap='gray')
fig.colorbar(fig_img, ax=axs[3],shrink=0.5)
axs[3].set_title('Canal Azul1')
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [10, 5]
size_imagen=100
R=np.ones((size_imagen,size_imagen))*255
G=np.ones((size_imagen,size_imagen))*128
B=np.ones((size_imagen,size_imagen))*0

imagen=np.dstack((R,G,B))
imagen=np.array(imagen,dtype=int)

plt.figure()
plt.imshow(imagen)
plt.title('imagen color:naranja')
plt.show()
```
