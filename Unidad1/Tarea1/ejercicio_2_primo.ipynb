# Ejercicio 2: Determinar si es Primo
Desarrolle una aplicación que determine si un número introducido por teclado es primo o no

```python
def es_primo(numero):
    if numero <= 1:
        return False
    elif numero == 2:
        return True
    elif numero % 2 == 0:
        return False
    else:
        for i in range(3, numero, 2):
            if numero % i == 0:
                return False
        return True
```

```python
numero = int(input("Ingrese un número entero: "))

if es_primo(numero):
    print(f"{numero} es un número primo")
else:
    print(f"{numero} NO es un número primo")
```
