# Ejercicio 7: Ordenamiento Burbuja
Desarrolle una aplicación que ordene números a través del método burbuja (Bubble Sort)

```python
def ordenamiento_burbuja(lista):
    longitud = len(lista)
    
    for i in range(longitud):
        for j in range(0, longitud - i - 1):
            if lista[j] > lista[j + 1]:
                auxiliar = lista[j]
                lista[j] = lista[j + 1]
                lista[j + 1] = auxiliar
    
    return lista
```

```python
def obtener_numeros():
    cantidad = int(input("¿Cuántos números desea ordenar? "))
    numeros = []
    for i in range(cantidad):
        numero = float(input(f"Ingrese el número {i + 1}: "))
        numeros.append(numero)
    return numeros
```

```python
numeros = obtener_numeros()

print(f"\nNúmeros originales: {numeros}")

numeros_copia = numeros.copy()
numeros_ordenados = ordenamiento_burbuja(numeros_copia)

print(f"Números ordenados: {numeros_ordenados}")
print(f"{'=' * 40}")
```

## Cómo funciona Bubble Sort:
1. Compara elementos adyacentes
2. Si están en orden incorrecto, los intercambia
3. Repite hasta que toda la lista esté ordenada

## Ejemplo:
- Original: [5, 2, 8, 1, 9]
- Ordenado: [1, 2, 5, 8, 9]
