# Ejercicio 5: Número Más Usado
Desarrolle una aplicación que permita introducir 7 números por teclado y determine cuál es el más usado y cuántas veces

```python
def obtener_siete_numeros():
    numeros = []
    for i in range(7):
        numero = int(input(f"Ingrese el número {i + 1}: "))
        numeros.append(numero)
    return numeros
```

```python
print("Ingrese 7 números:\n")
numeros = obtener_siete_numeros()

numero_mas_usado = numeros[0]
veces_maximo = numeros.count(numero_mas_usado)

for numero in numeros:
    veces_actual = numeros.count(numero)
    if veces_actual > veces_maximo:
        numero_mas_usado = numero
        veces_maximo = veces_actual

print(f"\n{'=' * 40}")
print(f"Números ingresados: {numeros}")
print(f"Número más usado: {numero_mas_usado}")
print(f"Veces que aparece: {veces_maximo}")
print(f"{'=' * 40}")
```

## Ejemplo:
- Números: [5, 3, 5, 8, 5, 2, 3]
- Número más usado: 5 (aparece 3 veces)
