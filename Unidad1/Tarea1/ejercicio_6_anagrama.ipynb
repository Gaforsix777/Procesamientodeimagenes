# Ejercicio 6: Anagrama
Desarrolle una aplicación que identifique si una palabra es un anagrama de otra (Bolo - Lobo)

```python
def es_anagrama(palabra1, palabra2):
    pal1_limpia = sorted(palabra1.replace(" ", "").lower())
    pal2_limpia = sorted(palabra2.replace(" ", "").lower())
    return pal1_limpia == pal2_limpia
```

```python
palabra1 = input("Ingrese la primera palabra: ")
palabra2 = input("Ingrese la segunda palabra: ")

if not palabra1.strip() or not palabra2.strip():
    print("Error: Debe ingresar palabras no vacías")
else:
    if es_anagrama(palabra1, palabra2):
        print(f"\n'{palabra1}' y '{palabra2}' SON anagramas")
    else:
        print(f"\n'{palabra1}' y '{palabra2}' NO son anagramas")
```

## Ejemplos de Anagramas:
- Bolo - Lobo
- amor - Roma
- listen - silent
- dormitory - dirty room
