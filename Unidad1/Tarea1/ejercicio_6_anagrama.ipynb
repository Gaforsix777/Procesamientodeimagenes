# Ejercicio 6: Análisis Avanzado de Anagramas
Aplicación profesional para identificar, validar y buscar anagramas con estadísticas y análisis.

```python
import re
from collections import Counter
from typing import List, Dict, Tuple

def normalizar_palabra(palabra: str) -> str:
    """
    Normaliza una palabra removiendo espacios, acentos y caracteres especiales.
    
    Args:
        palabra (str): Palabra a normalizar
        
    Returns:
        str: Palabra normalizada en minúsculas
    """
    return re.sub(r'[^a-záéíóú]', '', palabra.lower().strip())

def es_anagrama(palabra1: str, palabra2: str) -> bool:
    """
    Verifica si dos palabras son anagramas.
    
    Args:
        palabra1 (str): Primera palabra
        palabra2 (str): Segunda palabra
        
    Returns:
        bool: True si son anagramas, False en caso contrario
    """
    if not palabra1.strip() or not palabra2.strip():
        return False
    
    p1_norm = normalizar_palabra(palabra1)
    p2_norm = normalizar_palabra(palabra2)
    
    return sorted(p1_norm) == sorted(p2_norm)

def analizar_palabra(palabra: str) -> Dict:
    """
    Analiza una palabra y retorna estadísticas detalladas.
    
    Args:
        palabra (str): Palabra a analizar
        
    Returns:
        dict: Diccionario con información de la palabra
    """
    norm = normalizar_palabra(palabra)
    return {
        'original': palabra,
        'normalizada': norm,
        'longitud': len(norm),
        'frecuencia_caracteres': dict(Counter(norm)),
        'caracteres_unicos': len(set(norm))
    }

def buscar_anagramas(palabra: str, lista_palabras: List[str]) -> List[str]:
    """
    Busca todos los anagramas de una palabra en una lista.
    
    Args:
        palabra (str): Palabra a buscar anagramas
        lista_palabras (list): Lista de palabras donde buscar
        
    Returns:
        list: Lista de anagramas encontrados
    """
    return [p for p in lista_palabras if es_anagrama(palabra, p) and normalizar_palabra(p) != normalizar_palabra(palabra)]

def validar_entrada(texto: str) -> Tuple[bool, str]:
    """
    Valida que la entrada contenga solo letras.
    
    Args:
        texto (str): Texto a validar
        
    Returns:
        tuple: (es_válido, mensaje_error)
    """
    if not texto.strip():
        return False, "Error: No puede estar vacío"
    elif not re.match(r'^[a-záéíóúA-ZÁÉÍÓÚ\s]+$', texto):
        return False, "Error: Solo se permiten letras"
    return True, ""
```

```python
print("=" * 60)
print("ANALIZADOR PROFESIONAL DE ANAGRAMAS".center(60))
print("=" * 60)

# Entrada y validación
palabra1 = input("\nIngrese la primera palabra: ")
validado1, error1 = validar_entrada(palabra1)

if not validado1:
    print(f"❌ {error1}")
else:
    palabra2 = input("Ingrese la segunda palabra: ")
    validado2, error2 = validar_entrada(palabra2)
    
    if not validado2:
        print(f"❌ {error2}")
    else:
        # Análisis de ambas palabras
        print("\n" + "=" * 60)
        print("ANÁLISIS COMPARATIVO".center(60))
        print("=" * 60)
        
        # Información de palabra 1
        info1 = analizar_palabra(palabra1)
        print(f"\n📊 PALABRA 1: '{palabra1}'")
        print(f"   Normalizada: {info1['normalizada']}")
        print(f"   Longitud: {info1['longitud']} caracteres")
        print(f"   Caracteres únicos: {info1['caracteres_unicos']}")
        print(f"   Frecuencia: {info1['frecuencia_caracteres']}")
        
        # Información de palabra 2
        info2 = analizar_palabra(palabra2)
        print(f"\n📊 PALABRA 2: '{palabra2}'")
        print(f"   Normalizada: {info2['normalizada']}")
        print(f"   Longitud: {info2['longitud']} caracteres")
        print(f"   Caracteres únicos: {info2['caracteres_unicos']}")
        print(f"   Frecuencia: {info2['frecuencia_caracteres']}")
        
        # Resultado de anagrama
        print("\n" + "-" * 60)
        resultado = es_anagrama(palabra1, palabra2)
        if resultado:
            print("✅ RESULTADO: ¡SÍ SON ANAGRAMAS!")
        else:
            print("❌ RESULTADO: NO son anagramas")
            
            # Análisis por qué no son anagramas
            if info1['longitud'] != info2['longitud']:
                print(f"   Razón: Longitudes diferentes ({info1['longitud']} vs {info2['longitud']})")
            else:
                diferencias = set(info1['frecuencia_caracteres'].items()) ^ set(info2['frecuencia_caracteres'].items())
                print(f"   Razón: Diferentes frecuencias de caracteres")
```

## Ejemplos de Anagramas (Prueba):
- **Bolo** ↔ **Lobo** (mismos caracteres, diferente orden)
- **amor** ↔ **Roma** (a,m,o,r = a,m,o,r)
- **listen** ↔ **silent** (l,i,s,t,e,n = e,i,l,n,s,t)
- **dormitory** ↔ **dirty room** (incluyendo espacios)

```python
# Ejemplo: Buscar anagramas en un diccionario
diccionario_ejemplo = [
    "bolo", "lobo", "amor", "roma", "listen", "silent", 
    "nota", "tona", "atlas", "salsa", "alcalde", "acalled",
    "cartas", "trascas", "pausa", "saupa"
]

print("\n" + "=" * 60)
print("BÚSQUEDA EN DICCIONARIO".center(60))
print("=" * 60)

palabra_buscar = input("\nIngrese una palabra para buscar sus anagramas: ")
validado, error = validar_entrada(palabra_buscar)

if validado:
    anagramas = buscar_anagramas(palabra_buscar, diccionario_ejemplo)
    
    print(f"\n🔍 Anagramas de '{palabra_buscar}' encontrados:")
    if anagramas:
        for anagrama in anagramas:
            print(f"   ✓ {anagrama}")
    else:
        print("   No se encontraron anagramas en el diccionario")
else:
    print(f"❌ {error}")
```

```python
# Suite de pruebas automatizadas
print("\n" + "=" * 60)
print("SUITE DE PRUEBAS AUTOMATIZADAS".center(60))
print("=" * 60)

casos_prueba = [
    ("bolo", "lobo", True),
    ("amor", "roma", True),
    ("listen", "silent", True),
    ("python", "thonpy", True),
    ("gato", "perro", False),
    ("casa", "caza", False),
    ("abc", "def", False),
]

print("\nEjecutando " + str(len(casos_prueba)) + " casos de prueba...\n")
exitosos = 0

for pal1, pal2, esperado in casos_prueba:
    resultado = es_anagrama(pal1, pal2)
    estado = "✅ PASS" if resultado == esperado else "❌ FAIL"
    print(f"{estado} | '{pal1}' ↔ '{pal2}' = {resultado} (esperado: {esperado})")
    if resultado == esperado:
        exitosos += 1

print(f"\nResultado final: {exitosos}/{len(casos_prueba)} pruebas pasadas ({'100%' if exitosos == len(casos_prueba) else str(int(exitosos/len(casos_prueba)*100)) + '%'})")
```
