# Ejercicio 3: Mayor y Menor
Desarrolle una aplicación que permita ingresar diferentes números por teclado con un límite definido inicialmente (no menor a 3 números) y determine cuál de ellos es el mayor y el menor

```python
def obtener_numeros(cantidad):
    numeros = []
    for i in range(cantidad):
        numero = float(input(f"Ingrese el número {i + 1}: "))
        numeros.append(numero)
    return numeros
```

```python
cantidad = int(input("¿Cuántos números desea ingresar? (mínimo 3): "))

if cantidad < 3:
    print("Error: Debe ingresar mínimo 3 números")
else:
    print(f"\nIngrese {cantidad} números:")
    numeros = obtener_numeros(cantidad)
    
    mayor = max(numeros)
    menor = min(numeros)
    promedio = sum(numeros) / len(numeros)
    
    print(f"\n{'=' * 40}")
    print(f"Números ingresados: {numeros}")
    print(f"Mayor: {mayor}")
    print(f"Menor: {menor}")
    print(f"Promedio: {promedio:.2f}")
    print(f"{'=' * 40}")
```

## Ejemplo:
- Números: [5, 2, 8, 1, 9]
- Mayor: 9
- Menor: 1
- Promedio: 5.0
